# List
- Lists are used to store multiple items in a single variable.
- Lists are created using square brackets

# Properties: -
## List Items
- List items are ordered, changeable, and allow duplicate values.
- List items are indexed, the first item has index [0], the second item has index [1] etc.

## Ordered
- When we say that lists are ordered, it means that the items have a defined order, and that order will not change.
- If you add new items to a list, the new items will be placed at the end of the list.

## Mutable(changeable)
- The list is mutable, meaning that we can change, add, and remove items in a list after it has been created.

## Allow Duplicates
- Since lists are indexed, lists can have items with the same value

# 7. Changing List of Items

## a. Change Item Value
- To change the value of a specific item, refer to the index number

```python
item_list = ["apple", "banana", "cherry"]
print("Original list : ",item_list)
item_list[1] = "watermelon" #replacing banana by watermelon
print("Updated list : ",item_list)
```

## b.Change a Range of Item Values
- To change the value of items within a specific range, define a list with the new values, and refer to the range of index numbers where you want to insert the new values

```python
#Change the values "banana" and "cherry" with the values "papaya" and grapes
item_list = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
print("Original list : ",item_list)
item_list[1:3] = ["papaya", "grapes"]
print("Updated list : ",item_list)
```

<b>If you insert more items than you replace, the new items will be inserted where you specified, and the remaining items will move accordingly

# 8. Add List Items

## a. Append Items
- To add an item to the end of the list, use the append() method

```python
list1 = ["apple", "banana", "cherry"]
print("Original list : ",list1)
list1.append("orange") #element append at the last
print("Updated List : ",list1)
```

```python
print("List 1 : ",list1)
list2 = ['mango','papaya']
print("List2 : ",list2)
list1.append(list2) #whole list is appended
print("After Appending list1 and list2 - Updated List1 : ",list1)
```

## b. Extend Items
- To append elements from another list to the current list, use the extend() method.

```python
list1 = ["apple", "banana", "cherry"]
list2 = ['mango','papaya']
print("List1 : ",list1)
print("List2 : ",list2)
list1.extend(list2) #list1 is extended
print("Updated List1 after extending list1 and list2 : ",list1)
```

```python
print("Original List1 : ",list1)
list1.extend("orange")
print("Updated List1 after list1 is extended : ",list1)
```

## c. Insert Items
- To insert a new list item, without replacing any of the existing values, we can use the insert() method

```python
list1 = ["apple", "banana", "cherry"]
list1.insert(2, "watermelon") #inserting new element as index 2
print(list1)
```

# 10. Sort Lists

## a. Sort List Ascending order
- List objects have a sort() method that will sort the list alphanumerically, ascending, by default

```python
#Sorting list of strings
list1 = ["orange", "mango", "kiwi", "pineapple", "banana"]
list1.sort() #default is ascending order
print(list1)
```

```python
#Soritng numerical list
list1 = [100, 50, 65, 82, 23]
list1.sort() #default is ascending order
print(list1)
```

```python
#Soritng alphanumericl
list1 = ['apple',100,'papaya' ,50, 'cherry',65, 'banana',82, 23]
list1.sort() # sort is not supportted on combination of lists with integer and string 
print(list1)
```

## b. Sorting List in Descending Order

```python
#Sorting list of strings
list1 = ["orange", "mango", "kiwi", "pineapple", "banana"]
list1.sort(reverse = True) #default is ascending order
print(list1)
```

```python
#Soritng numerical list
list1 = [100, 50, 65, 82, 23]
list1.sort(reverse = True) #default is ascending order
print(list1)
```

## c. Customize Sort Function
- You can also customize your own function by using the keyword argument key = function.

- The function will return a number that will be used to sort the list (the lowest number first)

```python
def myfunc(n):
    return abs(n - 50)

list1 = [100, 50, 65, 82, 23]
list1.sort(key = myfunc) #keys -[50,0,15,32,abs(-27)] => [50,0,15,32,27]
print(list1)
```

## d. Case Insensitive Sort
- By default the sort() method is case sensitive, resulting in all capital letters being sorted before lower case letters

```python
#Case sensitive sorting can give an unexpected result
list1 = ["banana", "Orange", "Kiwi", "cherry"]
list1.sort()
print(list1)
```

```python
#Perform a case-insensitive sort of the list
list1 = ["banana", "Orange", "Kiwi", "cherry"]
list1.sort(key = str.lower)
print(list1)
```

## e. Reverse Order
- What if you want to reverse the order of a list, regardless of the alphabet?

- The reverse() method reverses the current sorting order of the elements.

```python
list1 = ["banana", "Orange", "Kiwi", "cherry"]
list1.reverse() 
print(list1)
```

# 11.Copy Lists
- You cannot copy a list simply by typing list2 = list1, because: list2 will only be a reference to list1, and changes made in list1 will automatically also be made in list2.

- There are ways to make a copy, one way is to use the built-in List method copy().

## a. Shallow Copy

```python
list1 = ["apple", "banana", "cherry"]
list2 = list1 #shallow copy
print("List1 : ",list1)
print("List2 : ",list2)
```

```python
#performing operation on list1
list1.append("papaya") #appending a new element in list1
print("After updating list1 : ", list1)
print("After updating list1, let us see what changed in list2 : ",list2)
print("You see that as we did shallow copy the changes we did in list1 got affected in list2 also.")
```

```python
#checking the identity of list1 and list2
print("List1 Identity : ", id(list1))
print("List2 Identity : ",id(list2))
```

<b> Both list1 and list2 shares same identity as new object was logically not created and both the lists are pointing to the same location

## b. Deep Copy

```python
list1 = ["apple", "banana", "cherry"]
list2 = list1.copy() #Deep copy
print("List1 : ",list1)
print("List2 : ",list2)
```

```python
#performing operation on list1
list1.append("papaya") #appending a new element in list1
print("After updating list1 : ", list1)
print("After updating list1, let us see what changed in list2 : ",list2)
print("You see that as we did Deep copy the changes we did in list1 does not affect list2.")
```

```python
#checking the identity of list1 and list2
print("List1 Identity : ", id(list1))
print("List2 Identity : ",id(list2))
```

<b> Both list1 and list2 shares different identity as new object was created and both the lists are pointing to the different location

# 13. Join Lists

## a. Join Two Lists
- There are several ways to join, or concatenate, two or more lists in Python.

- One of the easiest ways are by using the + operator.

```python
#Example-1
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
```

```python
#Example-2
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

for x in list2:
    list1.append(x)

print(list1)
```

```python
#Use the extend() method to add list2 at the end of list1
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]
list1.extend(list2)
list1
```
